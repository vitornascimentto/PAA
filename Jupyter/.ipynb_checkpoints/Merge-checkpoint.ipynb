```python
from random import randrange
import matplotlib.pyplot as plt
import time
```

```python
def mergeSort(lista):

    if len(lista) > 1:

        meio = len(lista)//2
        #também é valido: meio = int(len(lista)/2)

        listaDaEsquerda = lista[:meio]
        listaDaDireita = lista[meio:]

        mergeSort(listaDaEsquerda)
        mergeSort(listaDaDireita)

        i = 0
        j = 0
        k = 0

        while i < len(listaDaEsquerda) and j < len(listaDaDireita):

            if listaDaEsquerda[i] < listaDaDireita[j]:
                lista[k]=listaDaEsquerda[i]
                i += 1
            else:
                lista[k]=listaDaDireita[j]
                j += 1
            k += 1

        while i < len(listaDaEsquerda):

            lista[k]=listaDaEsquerda[i]
            i += 1
            k += 1

        while j < len(listaDaDireita):
            lista[k]=listaDaDireita[j]
            j += 1
            k += 1

```

```python
def num(tamanho, ordenacao):
    nums = [randrange(1000) for x in range(tamanho)]
    
    if ordenacao == 'Crescente':
        return sorted(nums)
    elif ordenacao == 'Decrescente':
        return sorted(nums, reverse=True)
    elif ordenacao == 'Desordenado':
        return nums
```

```python
var1 = num(100, 'Desordenado')
var2 = num(100, 'Crescente')
var3 = num(100, 'Decrescente')
```

```python
def graphic(variavel):
    plt.title('Messy Numbers')
    plt.grid(True)
    plt.ylabel('Numbers')
    plt.plot(variavel)
    plt.show()
    plt.clf()
```

```python
graphic(var1)

inicio = time.time()

mergeSort(var1)

fim = time.time()

print('Tempo:' ,(fim - inicio))

graphic(var1)
```

```python
graphic(var2)

inicio = time.time()

mergeSort(var2)

fim = time.time()

print('Tempo:' ,(fim - inicio))

graphic(var2)
```

```python
graphic(var3)

inicio = time.time()
mergeSort(var3)
fim = time.time()

print('Tempo:' ,(fim - inicio))

graphic(var3)
```

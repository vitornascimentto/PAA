```python
from random import randrange
import matplotlib.pyplot as plt
import time
```

```python
def particao(a, ini, fim):
    pivo = a[fim-1]
    start = ini
    end = ini
    for i in range(ini,fim):
        if a[i] > pivo:
            end += 1
        else:
            end += 1       
            start += 1
            aux = a[start-1]
            a[start-1] = a[i]
            a[i] = aux
    return start-1
    
def quickSort(a, ini, fim):
    if ini < fim:
        pp = randparticao(a, ini, fim)
        quickSort(a, ini, pp)
        quickSort(a, pp+1,fim)
    return a
    
def randparticao(a,ini,fim):
    rand = random.randrange(ini,fim)
    aux = a[fim-1]
    a[fim-1] = a[rand]
    a[rand] = aux
    return particao(a,ini,fim)
```

```python
def num(tamanho, ordenacao):
    nums = [randrange(1000) for x in range(tamanho)]
    
    if ordenacao == 'Crescente':
        return sorted(nums)
    elif ordenacao == 'Decrescente':
        return sorted(nums, reverse=True)
    elif ordenacao == 'Desordenado':
        return nums
```

```python
def graphic(variavel):
    plt.title('Messy Numbers')
    plt.grid(True)
    plt.ylabel('Numbers')
    plt.plot(variavel)
    plt.show()
    plt.clf()
```

```python
var1 = num(100, 'Desordenado')
var2 = num(100, 'Crescente')
var3 = num(100, 'Decrescente')
```

```python
graphic(var1)

inicio = time.time()

quickSort(var1, 0, len(var1))

fim = time.time()

print('Tempo:' ,(fim - inicio))

graphic(var1)
```

```python
# Imports
from random import randrange
import matplotlib.pyplot as plt
import time
```

```python
# Sorting function
def gnomeSort(array):
    
    inicio = time.time()
    
    index = 0
    
    contComparacoes = 0
    contSwaps = 0
    
    while index < len(array):
        if index == 0 or array[index - 1] <= array[index]:
            index += 1
            contComparacoes += 1
        else:
            array[index], array[index - 1] = array[index - 1], array[index]
            index -= 1
            contSwaps += 1
            
    print('Comparações:' ,contComparacoes)
    print('Swaps:' ,contSwaps)
    
    fim = time.time()
    
    print('Tempo:', (fim - inicio))
```

```python
# Display function
def show(array):
  for i in array:
    print(i, end=' ')
  
  print()
```

```python
def num(tamanho, ordenacao):
    nums = [randrange(1000) for x in range(tamanho)]
    
    if ordenacao == 'Crescente':
        return sorted(nums)
    elif ordenacao == 'Decrescente':
        return sorted(nums, reverse=True)
    elif ordenacao == 'Desordenado':
        return nums
```

```python
# Creating an Array with Random Numbers
#nums = [randrange(1000) for x in range(1000)]

var1 = num(100, 'Desordenado')
var2 = num(100, 'Crescente')
var3 = num(100, 'Decrescente')
```

```python
# Messy array
print ("Numbers:")
show(var1)
```

```python
def graphic(variavel):
    plt.title('Messy Numbers')
    plt.grid(True)
    plt.ylabel('Numbers')
    plt.plot(variavel)
    plt.show()
    plt.clf()
```

```python
# Graphic
graphic(var1)
```

```python
# Calling the sorting function
gnomeSort(var1)
```

```python
# Ordered array
print ("Sorted numbers:")
show(var1)
```

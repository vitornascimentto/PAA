```python
from random import randrange
import matplotlib.pyplot as plt
import time
```

```python
def insertion_sort( lista ):
    cont = 0
    for i in range( 1, len( lista ) ):
        chave = lista[i]
        k = i
        cont += 1
        while k > 0 and chave < lista[k - 1]:
            cont+=1
            lista[k] = lista[k - 1]
            k -= 1
        lista[k] = chave
        
    print (cont)
```

```python
def num(tamanho, ordenacao):
    nums = [randrange(1000) for x in range(tamanho)]
    
    if ordenacao == 'Crescente':
        return sorted(nums)
    elif ordenacao == 'Decrescente':
        return sorted(nums, reverse=True)
    elif ordenacao == 'Desordenado':
        return nums
```

```python
def graphic(variavel):
    plt.title('Messy Numbers')
    plt.grid(True)
    plt.ylabel('Numbers')
    plt.plot(variavel)
    plt.show()
    plt.clf()
```

```python
var1 = num(100, 'Desordenado')
var2 = num(100, 'Crescente')
var3 = num(100, 'Decrescente')
```

```python
graphic(var1)

inicio = time.time()

insertion_sort(var1)

fim = time.time()

print('Tempo:' ,(fim - inicio))

graphic(var1)
```

```python
graphic(var2)

inicio = time.time()

insertion_sort(var2)

fim = time.time()

print('Tempo:' ,(fim - inicio))

graphic(var2)
```

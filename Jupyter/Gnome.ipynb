```python
# Imports
from random import randrange
import matplotlib.pyplot as plt
```

```python
# Sorting function
def gnomeSort(array):
    index = 0
    
    contComparacoes = 0
    contSwaps = 0
    
    while index < len(array):
        if index == 0 or array[index - 1] <= array[index]:
            index += 1
            contComparacoes += 1
        else:
            array[index], array[index - 1] = array[index - 1], array[index]
            index -= 1
            contSwaps += 1
            
    print('Comparações:' ,contComparacoes)
    print('Swaps:' ,contSwaps)
```

```python
# Display function
def show(array):
  for i in array:
    print(i, end=' ')
  
  print()
```

```python
# Creating an Array with Random Numbers
nums = [randrange(200) for x in range(100)]
```

```python
# Messy array
print ("Numbers:")
show(nums)
```

```python
# Graphic
plt.title('Messy Numbers')
plt.grid(True)
plt.ylabel('Numbers')
plt.plot(nums)
plt.show()
plt.clf()
```

```python
# Calling the sorting function
gnomeSort(nums)
```

```python
# Ordered array
print ("Sorted numbers:")
show(nums)
```

```python
# Graphic
plt.title('Ordered Numbers')
plt.grid(True)
plt.ylabel('Numbers')
plt.plot(nums)
plt.show()
```
